```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, confusion_matrix
from sklearn.model_selection import KFold, cross_val_score
import joblib

import seaborn as snsfrom 
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\bus_dataset.csv")
df.head()  # Display the first 5 rows


```

```python
df['hour'] = df['time'].str.split(':').str[0].astype(int)
df['minute'] = df['time'].str.split(':').str[1].astype(int)




    
```

```python
df = df.drop('time', axis=1)
```

```python
df.head()
```

```python


df['Outcome'] = np.random.randint(0, 2, len(df))
```

```python
x = df.drop("Outcome", axis=1) 
y = df["Outcome"]               

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
print("features for training",x_train.shape)
print("features for training",x_test.shape)
```

```python
lr=LogisticRegression(max_iter=500)
lr.fit(x_train,y_train)
```

```python
pred_lr=lr.predict(x_test)
```

```python
print("accuracy", accuracy_score(y_test,pred_lr))
print("classification", classification_report(y_test,pred_lr))
print("confusion matrix", confusion_matrix(y_test,pred_lr))


      
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
pred_dt=dt.predict(x_test)
```

```python
print("accuracy", accuracy_score(y_test,pred_dt))
print("classification", classification_report(y_test,pred_dt))
print("confusion matrix", confusion_matrix(y_test,pred_dt))
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
```

```python
pred_kn=knn.predict(x_test)
print(" knn accuracy", accuracy_score(y_test,pred_kn))
```
